```python
import nltk
import re
import string
from nltk.corpus import stopwords
nltk.download('punkt')
nltk.download('stopwords')
from nltk.tokenize import word_tokenize
```

```python
df['synopsis'] = df['synopsis'].astype(str)
```

```python
stop_words = stopwords.words()

def cleaning(synopsis):        
    # converting to lowercase, removing URL links, special characters, punctuations...
    synopsis = synopsis.lower()
    synopsis = re.sub('<.*?>+', '', synopsis)
    synopsis = re.sub('[%s]' % re.escape(string.punctuation), '', synopsis)
    synopsis = re.sub('\n', '', synopsis)
    synopsis = re.sub('[’“”…]', '', synopsis)
    # removing the stop-words  
    text_tokens = word_tokenize(synopsis)
    tokens_without_sw = [word for word in text_tokens if not word in stop_words]
    filtered_sentence = (" ").join(tokens_without_sw)
    synopsis = filtered_sentence
    
    return synopsis

dt = df['synopsis'].apply(cleaning)

from collections import Counter
p = Counter(" ".join(dt).split()).most_common(10)
rslt = pd.df(p, columns=['Word', 'Frequency'])
print(rslt)
```
